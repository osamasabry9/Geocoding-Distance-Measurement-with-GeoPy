# 3. Load & Explore the Data

```python
import pandas as pd

# Load the uploaded CSV file (example: 'ae_addresses.csv')
df = pd.read_csv('ae_addresses.csv')  # Adjust the filename as needed

# Preview the data
df[['city', 'lat', 'lng', 'admin_name']].head()
```

# 4. Geocode Sample City with geopy

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

# Initialize geolocator with user agent
geolocator = Nominatim(user_agent="uae_location_app")

# Add rate limiter to avoid query blocking
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

# Example: geocode a sample city
location = geocode("Jebel Ali, Dubai, UAE")
print(location.address, location.latitude, location.longitude)
```

**4.2 Robust Function with Error Handling**

```python
def get_location_info(address):
    try:
        location = geocode(address)
        if location:
            print(f"📍 Full address: {location.address}")
            print(f"🧭 Latitude: {location.latitude}")
            print(f"🧭 Longitude: {location.longitude}")
            print(f"📌 Raw data: {location.raw}")
        else:
            print("❌ Location not found.")
    except Exception as e:
        print(f"⚠️ Error during geocoding: {e}")


get_location_info("Ajman University, UAE")
```

**4.3: Batch Geocoding Multiple Addresses**


```python
addresses = [
    "Burj Khalifa, Dubai",
    "Sheikh Zayed Grand Mosque, Abu Dhabi",
    "Sharjah University",
    "Ajman Museum"
]

results = []
for addr in addresses:
    loc = geocode(addr)
    results.append({
        "address": addr,
        "latitude": loc.latitude if loc else None,
        "longitude": loc.longitude if loc else None,
        "full_address": loc.address if loc else "Not found"
    })

import pandas as pd
df_geo = pd.DataFrame(results)
df_geo
```

# 5. Plot UAE Cities on Interactive Map (Folium)

```python
import folium

# Create a base map centered around the average coordinates
m = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], zoom_start=7)

# Add markers for each city
for _, row in df.iterrows():
    folium.Marker(
        location=[row['lat'], row['lng']],
        popup=f"{row['city']}, {row['admin_name']}",
        tooltip=row['city']
    ).add_to(m)

# Save and show the map
m.save('uae_map.html')
m
```

# 6. Reverse Geocoding: Coordinates → Address

```python
# Add reverse geocoding function
reverse = RateLimiter(geolocator.reverse, min_delay_seconds=1)

# Test with known coordinates
point = (25.1972, 55.2744)  # Burj Khalifa
address = reverse(point)
print("Reverse geocoded address:", address.address)
```

# 7. Distance Calculation Between Two Cities


```python
from geopy.distance import geodesic

# Define two cities
city1 = df[df['city'] == 'Dubai'][['lat', 'lng']].values[0]
city2 = df[df['city'] == 'Abu Dhabi'][['lat', 'lng']].values[0]

# Calculate the geodesic distance in kilometers
distance_km = geodesic(city1, city2).km
print(f"Distance between Dubai and Abu Dhabi: {distance_km:.2f} km")
```
